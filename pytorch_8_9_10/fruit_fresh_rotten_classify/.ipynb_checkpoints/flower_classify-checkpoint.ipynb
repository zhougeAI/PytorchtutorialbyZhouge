```python
import torch
import torch.nn as nn
import torchvision
import numpy
import matplotlib.pyplot as plt
from torchsummary import summary
```

```python
# 定义模型类
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = torchvision.models.vgg16(pretrained=True)
        self.classifier = nn.Sequential(nn.Linear(25088,100),
                                       nn.ReLU(),
                                       nn.Dropout(p=0.5),
                                       nn.Linear(100,5))
        for param in self.model.parameters():
            param.requires_grad = False
        
        self.model.classifier = self.classifier
    def forward(self,input):
        output = self.model(input)
        return  output
```

```python
# 定义模型，实例化类
mymodel = MyModel()
```

```python
summary(mymodel,input_size=[(3,50,50)],batch_size=1,device='cpu')
```

```python
# 数据集的读取
# 数据集的图片读进来，并且做一些预处理，转化为张量
# torchvision.datasets.ImageFolder() 从文件夹读取数据集
# 训练集
train_dataset = torchvision.datasets.ImageFolder( root='./flowers/train/',
                                          transform=torchvision.transforms.Compose([torchvision.transforms.ToTensor(),
                                                      torchvision.transforms.RandomHorizontalFlip(),
                                                      torchvision.transforms.RandomRotation((20)),
                                                      torchvision.transforms.Resize((50,50)),
                                                      ])
                                          )
# 验证集
val_dataset = torchvision.datasets.ImageFolder( root='./flowers/val/',
                                          transform=torchvision.transforms.Compose([torchvision.transforms.ToTensor(),
                                                      torchvision.transforms.Resize((50,50)),
                                                      ])
                                          )

# 数据加载器
# 训练集
train_dataloader = torch.utils.data.DataLoader(dataset=train_dataset,batch_size=32,shuffle=True)
# 验证集
val_dataloader = torch.utils.data.DataLoader(dataset=val_dataset,batch_size=32,shuffle=True)
```

```python
# 可视化导入器
import matplotlib.pyplot as plt
import numpy as np
data_iter = iter(train_dataloader)
name = [ 'daisy','dandelion','rose','sunflower','tulip']
for i in range(32):
    a,label = data_iter.__next__()
    show_image = np.transpose(a[0],(1,2,0))
    plt.subplot(4,8,i+1)
    plt.imshow(show_image)
    plt.title(name[label[0].item()])
    plt.axis('off')
plt.show()
```

```python
# 优化器
optimizer = torch.optim.Adam(mymodel.parameters(),lr=0.001)
```

```python
def get_infinite_dataset(dataloader):
    while True:
        for item in dataloader:
            yield item
```

```python
# 训练和验证网络
data_loader = get_infinite_dataset(train_dataloader)
total_epoch = 1000
# 使用 GPU 做训练,把模型搬到GPU上
mymodel.to('cuda:0')

for i in range(total_epoch+1):
    # 训练步骤
    mymodel.train()  #将模型调整为训练模式
    input,label = data_loader.__next__()
    # 把数据搬到GPU上
    input = input.to('cuda:0')
    label = label.to('cuda:0')
    output = mymodel(input) # 计算模型输出
    optimizer.zero_grad() #清空优化器的历史梯度
    loss =  nn.CrossEntropyLoss()(output,label)
    loss.backward()   # 求损失函数的梯度并反向传播
    optimizer.step()  # 优化器优化一次模型参数
    if i % 100 == 0:
        print('epoch:',i,'loss:',loss.item())
    
    if i % 200 == 0:
        # 验证
        mymodel.eval()
        total_correct = 0
        total_result = 0
        for item in val_dataloader:
            # 把数据搬到GPU上
            val_input,val_label = item
            val_input = val_input.cuda()
            val_label = val_label.cuda()
            output = mymodel(val_input)  # 计算输出
            _,pred = torch.max(output,1)  # 把预测的标签最大值那一类取出
            # 统计计算了多少个样本
            total_result = total_result + val_label.size(0)
            # 统计预测正确了多少个样本
            total_correct = total_correct + ( pred == val_label ).sum()
            # 总的正确率
            total_accuracy = total_correct / total_result 
        # 打印信息
        print('epoch:',i,' ','total accuracy:',total_accuracy.item())
```
